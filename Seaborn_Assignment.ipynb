### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

###### Seaborn is a popular data visualization library built on top of Matplotlib, that provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots that can be created using Seaborn and their respective uses:

###### Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. They are commonly used in exploratory data analysis to identify patterns and outliers in data.

###### Histogram Plot: Histograms are used to visualize the distribution of a continuous variable. They are commonly used to identify the shape, center, and spread of a dataset.

###### Line Plot: Line plots are used to visualize the trend of a continuous variable over time or another continuous variable. They are commonly used in time-series analysis or to visualize the relationship between two continuous variables.

###### Bar Plot: Bar plots are used to visualize the distribution of a categorical variable. They are commonly used to compare the frequencies or proportions of different categories.

###### Heatmap Plot: Heatmaps are used to visualize the relationship between two categorical variables. They are commonly used in correlation analysis or to identify patterns in large datasets.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
###### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot with different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
###### Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create box plots for age and fare, grouped by pclass
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram for the price column, with bars colored by cut
sns.histplot(data=diamonds, x="price", hue="cut")
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with markers colored by species
sns.pairplot(data=iris, hue="species")
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Reshape the data to a pivot table
flights = flights.pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(data=flights, annot=True, fmt="d", cmap="YlGnBu")
```
